# Tutorial sull'uso dei dati del Servizio Copernicus per il Cambiamento del Clima (C3S)

In questo tutorial vedremo come accedere ed elaborare i dati del **Copernicus Climate Change Service (C3S)**. Il servizio relativo ai cambiamenti climatici di Copernicus (C3S) che fornisce informazioni autorevoli sul clima passato, presente e futuro. Il C3S è affidato al Centro Europeo per le Previsioni Meteorologiche a Medio Termine (ECMWF) dalla Commissione Europea.

Forniamo informazioni su una serie di tematiche e ambiti settoriali attraverso un archivio di dati climatici (CDS). Il CDS è concepito per consentire agli utenti di personalizzare i servizi in base a esigenze pubbliche o commerciali più specifiche. Principalmente forniamo i seguenti dati:

- **Essential Climate Variables (ECVs)** misurate tramite sensori satellitari e terrestri.
- **Reanalysis** combina modelli con i dati osservazionali dei satelliti e dei sensori terrestri per ottenere la migliore comprensione possibile del clima del passato.
- **Seasonal forecasts** previsioni stagionali.
- **Climate projections** proiezioni a lungo termine, seconda vari scenari.
- **Climate indices** indici composti da diversi input climatici.

### Creare mappe con i dati sis-urban-climate-cities
#### Scaricare dati
Nel CDS https://cds.climate.copernicus.eu/ cerca il dataset `Climate variables for cities in Europe from 2008 to 2017`. Poi, nella finestra di `Download data`, scegli i seguenti parametri:

- Variable: `air_temperature`
- City: `milan`
- Year: `2017`
- Month: `01`
- Format: `zip`

In fondo alla pagina, scegli `Show API request`. Poi copia il codice, e incollalo dentro la cella seguente, cambiando solo il nome del file .zip alla fine del blocco di codici. Poi lancialo per scaricare i dati...

```python
import cdsapi

c = cdsapi.Client()

c.retrieve(
    'sis-urban-climate-cities',
    {
        'variable': 'air_temperature',
        'city': 'milan',
        'year': '2017',
        'month': '01',
        'format': 'zip',
    },
    './milan/urban_climate_at.zip')
```

Ora fai la stessa cosa, ma per la variabile `wind speed`:

```python
c.retrieve(
    'sis-urban-climate-cities',
    {
        'variable': 'wind_speed',
        'city': 'milan',
        'year': '2017',
        'month': '01',
        'day': '15',
        'format': 'zip',
    },
    './milan/urban_climate_ws.zip')
```

Una volta scaricati i dati, devi estrarre i file zip:

```python
# Funzione per estrarre i file zip
def zipextract(zip_file):
    with ZipFile(zip_file, 'r') as zipObj: # Crea un Oggetto Zipfile e inseriscici il file zip
        zipObj.extractall('milan') # Estrai tutti i contenuti del file zip nella cartella di default
```

```python
at_zip = './milan/urban_climate_at.zip'
ws_zip = './milan/urban_climate_ws.zip'
zipextract(at_zip)
zipextract(ws_zip)
```

Poi puoi leggere i file NetCDF in un 'Dataset' di Xarray:

```python
at_nc = './milan/tas_Milan_UrbClim_2017_01_v1.0.nc'
ws_nc = './milan/sfcWind_Milan_UrbClim_2017_01_v1.0.nc'
at_ds = xr.open_dataset(at_nc)
ws_ds = xr.open_dataset(ws_nc)
```

Esamina il contenuto del dataset:

```python
at_ds
```

```python
ws_ds
```

Per lavorare meglio con i parametri singoli, devi convertire i `Xarray Dataset` in `Xarray Data Array`:

```python
at_da = at_ds['tas']
ws_da = ws_ds['sfcWind']
```

Per vedere le mappe dei vari parametri, rendi monodimensionale la variabile del tempo:

```python
at_mean = at_da.mean(dim='time')
ws_mean = ws_da.mean(dim='time')
```

Ora puoi creare mappe per visualizzare i dati:

```python
def visualise(variable, longitude, latitude, colourmap, title, label):
    """ 
    Visualizza un Xarray Data Array con la funzione di matplotlib 'pcolormesh'.
    
    Parametri:
        variable: variabile dei dati, per esempio la temperatura dell'aria
        longitude: valori di longitudine per la variabile di dati
        latitude: valori di latitudine per la variabile di dati
        colourmap: mappa di colori per mostrare la gamma di valori
        title: titolo della mappa
        label: etichetta nella legenda, ad esempio 'Temperatura (Kelvin)'
    """
    # creare il pannello della figura 
    fig = plt.figure(figsize=(10,5))
    # creare la mappa usando la proiezione di PlateCarree
    ax = plt.subplot(1,1,1, projection=ccrs.PlateCarree())
    # aggiungere una griglia lat lon
    gl = ax.gridlines(draw_labels=True, linewidth=1, color='gray', alpha=0.5, linestyle='--')
    gl.right_labels = None
    gl.top_labels = None
    # aggiungere un titolo
    ax.set_title(title)
    # creare il grafico
    im = plt.pcolormesh(longitude, latitude,
                        variable, cmap=colourmap, transform=ccrs.PlateCarree())
    # aggiungere la legenda
    cbar = plt.colorbar(im)
    cbar.set_label(label)

    return fig, ax
```

```python
visualise(at_mean, at_mean.longitude, at_mean.latitude, 'bwr', 
          'Surface Temperature Milan January 2017', 'Surface Temperature (Kelvin)')
```

```python
visualise(ws_mean, ws_mean.longitude, ws_mean.latitude, 'viridis', 
          'Wind Speed Milan January 2017', 'Wind Speed (m/s)')
```

#### Calcolare indice di wind chill

Qui vediamo l'esempio di un indice creato da vari parametri. In questo caso l'indice di wind chill, che si può calcolare con la temperatura dell'aria e la velocità del vento, tutte due vicino alla superficie. Esistono vari indici per calcolare il wind chill, noi useremo l'indice del Nord America e del Regno Unito:

$\textit{T}_{WC} = 13.12 + 0.6215\textit{T}_{a} - 11.37\upsilon^{0.16} + 0.3965\textit{T}_{a}\upsilon^{0.16}$

where:
- $\textit{T}_{WC}$ = indice di wind chill
- $\textit{T}_{a}$ = temperature dell'aria in Celsius
- $\upsilon$ = velocita del vento a 10 m altezza di un anemometro standard, in chilometri all'ora.

Per calcolare $\textit{T}_{WC}$ devi prima assicurarti che i dati siano nelle unità giuste. Per la velocità del vento devi convertire da m/s a km/h, e per la temperatura dell'aria da Kelvin a Celsius:

```python
# velocità del vento, convertire da m/s a km/h: si10 * 1000 / (60*60)
ws_kmh = ws_mean * 3600 / 1000
# temperatura dell'aria, convertire da Kelvin a Celsius: t2m - 273.15
at_cel = at_mean - 273.15
```

Poi puoi applicare la formula per calcolare il wind chill:

```python
twc = 13.12 + (0.6215*at_cel) - (11.37*(ws_kmh**0.16)) + (0.3965*at_cel*(ws_kmh**0.16))
```

Alla fine puoi creare una mappa del wind chill a gennaio 2017 a Milano:

```python
visualise(twc, twc.longitude, twc.latitude, 'viridis', 
          'Wind Chill Milan January 2017', 'Wind Chill Index')
```

### Calcolare anomalie e climatologie per Milano con dati ERA-5
#### Scaricare dati
Nel CDS https://cds.climate.copernicus.eu/ cerca il dataset `ERA5-Land monthly averaged data from 1981 to present`. Poi, nella finestra `Download data`, scegli i seguenti parametri:

- Product type: `Monthly averaged reanalysis`
- Variable: `2m temperature`
- Year: `1981 to 2019`
- Month: `all`
- Time: `00:00` (default)
- Geographical area: North: `45.5`, East: `9.3`, South: `45.4`, West: `9.1` 
- Format: `NetCDF`

In fondo alla pagina, scegli `Show API request`. Poi copia il codice, e incollalo dentro la cella seguente, cambiando solo il nome del file .zip alla fine del blocco di codici. Poi lancialo per scaricare i dati...

```python
import cdsapi

c = cdsapi.Client()

c.retrieve(
    'reanalysis-era5-land-monthly-means',
    {
        'format': 'netcdf',
        'product_type': 'monthly_averaged_reanalysis',
        'variable': '2m_temperature',
        'year': [
            '1981', '1982', '1983',
            '1984', '1985', '1986',
            '1987', '1988', '1989',
            '1990', '1991', '1992',
            '1993', '1994', '1995',
            '1996', '1997', '1998',
            '1999', '2000', '2001',
            '2002', '2003', '2004',
            '2005', '2006', '2007',
            '2008', '2009', '2010',
            '2011', '2012', '2013',
            '2014', '2015', '2016',
            '2017', '2018', '2019',
        ],
        'month': [
            '01', '02', '03',
            '04', '05', '06',
            '07', '08', '09',
            '10', '11', '12',
        ],
        'time': '00:00',
        'area': [
            45.5, 9.1, 45.4,
            9.3,
        ],
    },
    './milan/ERA5_Land_t2m_Milan.nc')
```

Segui gli stessi passi di prima per creare un Xarray Dataset dei dati, poi un Xarray Data Array per la variabile `2m temperature` o `t2m` come il nome dato nel file NetCDF.

```python
t2m_nc = './milan/ERA5_Land_t2m_Milan.nc'
t2m_ds = xr.open_dataset(t2m_nc)
```

```python
t2m_ds
```

```python
t2m_da = t2m_ds['t2m']
```

Questi dati hanno una risoluzione spaziale molto inferiore rispetto a quella del sis-urban-climate-cities. Però in questo caso non vuoi creare una mappa, ma lavorare con una serie temporale su un punto, cioè un unico "pixel" su Milano. Quindi devi rendere monodimensionale la variabile dello spazio:

```python
t2m = t2m_da.mean(["longitude", "latitude"])
```

#### Fare grafici di serie temporali: trend e anomalie
Ora puoi vedere come cambia la temperatura di Milano nella serie temporale dal 1981 al 2019.

```python
fig = plt.figure(figsize=(8,5))
ax = plt.subplot()
ax.set_ylabel('t2m')
ax.set_xlabel('time')
ax.plot(t2m.time, t2m, color='green', label='2m Temperature Milan')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
ax.set_title('2m Temperature in Milan from 1981 to 2019')
```

In questo grafico è difficile notare tendenze della temperatura, perché c'è una differenza stagionale troppo grande per percepire differenze sottili. Per rimuovere le variazioni stagionali, esamina la media annuale:

```python
t2m_y = t2m.groupby('time.year').mean('time')
```

Inoltre, invece di vedere le temperature assolute, analizza le anomalie rispetto a una media. Per farlo, devi prima calcolare la media per un periodo di riferimento. Un periodo usato frequentemente è quello dal 1981 al 2010:

```python
ref = t2m_y.where((t2m_y.year > 1980) & (t2m_y.year < 2011), drop=True)
ref = ref.mean(dim="year")
```

```python
anom = t2m_y - ref
```

```python
# Creare una linea orizzontale tratteggiata per indicare la media dell'anomalia
mean_line = xr.DataArray(0.0, coords=[('year', np.arange(1981,2010))])

fig = plt.figure(figsize=(8,5))
ax = plt.subplot()
ax.set_ylabel('t2m anomaly')
ax.set_xlabel('year')
ax.plot(anom.year, anom, color='green', label='2m Temperature Anomalies Milan')
ax.plot(mean_line.year, mean_line, color='red', linestyle='dashed', label='Mean anomaly 1981-2010')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
ax.set_title('2m Temperature Anomalies in Milan from 1981 to 2019')
```

Ora vedrai più chiaramente la tendenza della temperatura dal 1980 al 2019.

#### Calcolare la climatologia
Qui farai il calcolo della climatologia mensile. In particolare, calcolerai la media e deviazione standard della temperatura nella seria temporale per ogni singolo mese. Poi farai un confronto fra le temperature mensili del 2019 e la climatologia.

Per capire meglio i valori della temperatura, convertila in Celsius:

```python
t2mC = t2m - 273.15
```

Calcola quindi la climatologia:

```python
t2m_m_m = t2mC.groupby('time.month').mean('time')
t2m_m_sd = t2mC.groupby('time.month').std('time')
```

E le temperature del 2019:

```python
t2m_2019 = t2mC.sel(time='2019')
```

Infine, puoi creare un grafico che fa un confronto fra le temperature mensili del 2019 e la climatologia.

```python
fig = plt.figure(figsize=(10,5))
ax = plt.subplot()
ax.set_ylabel('t2m (Celsius)')
ax.set_xlabel('month')
ax.plot(t2m_m_m.month, t2m_m_m, color='gray', linestyle='dashed', label='t2m mean, shading: +/- SD')
ax.plot(t2m_m_m.month, (t2m_m_m + t2m_m_sd), color='gray', alpha=0.1)
ax.plot(t2m_m_m.month, (t2m_m_m - t2m_m_sd), color='gray', alpha=0.1)
ax.fill_between(t2m_m_m.month, (t2m_m_m + t2m_m_sd), (t2m_m_m - t2m_m_sd), color='gray', alpha=0.1)
ax.plot(t2m_m_m.month, t2m_2019, color='red', label='t2m 2019')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
ax.set_title('t2m climatology (1981 to 2019) compared with 2019')
```
